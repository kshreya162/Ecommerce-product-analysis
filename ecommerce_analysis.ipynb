```python
import pandas as pd
```

```python
data={
    "product": ["shoes", "tshirt","phone", "laptop","watch", "bag", "tablet"],
    "category":["fashion", "fashion","electronics","electronics","accessories","fashion","electronics"],
    "price":[2000, 800, 15000, 55000,3000,1200,20000],
    "rating": [4.5,4.2, 4.7, 4.8, 4.3, 4.1, 4.6],
    "stock":[20,50,15,8,30,40,10]
}
df=pd.DataFrame(data)
print(df)
```

```python
print(df.head())
print(df.shape)
print(df.columns)
df.info()
```

```python
print("Average price of the products: ",df["price"].mean())
print("Highest price product: ",df[df["price"]==df["price"].max()])
print("Products with price greater than 10000:\n",df[df["price"]>10000])
df[["product","price"]]
```

```python
print(df.groupby("category")["price"].mean())
print(df.groupby("category")["product"].count())
print(df.groupby("category")["rating"].mean())
df.groupby("category")["stock"].sum()
```

```python
print(df.sort_values("price", ascending=False))
print(df.sort_values("rating",ascending=False).head(3))
df.loc[df["price"].idxmin()]
```

```python
print("Most expensive product:",df.loc[df["price"].idxmax(),"product"])
print("Cheapest product:",df.loc[df["price"].idxmin(),"product"])
print("Best rated product:", df.loc[df["rating"].idxmax(),"product"])
print("category with highest average price:",df.groupby("category")["price"].mean().idxmax())
print("Category with highest average rating:",df.groupby("category")["rating"].mean().idxmax())
```

```python
print("Total stock available:",df["stock"].sum())
print("Product with lowest stock:\n",df.loc[df["stock"].idxmin(),"product"])
print("Product with stock less than 15:\n",df[df["stock"]<15])
print("Average stock per category:\n",df.groupby("category")["stock"].mean())
```

```python
print("Product priced above average price:\n",df[df["price"]>df["price"].mean()])
print("Products with rating greater than 4.5:\n",df[df["rating"]>4.5])
print("Electronics products only:\n",df[df["category"] == "electronics"])
print("Fashion products priced above 1000:\n",df[(df["category"]=="fashion") & (df["price"]>1000)])
```

```python
print("Top 3 expensive products:\n",df.sort_values("price",ascending=False).head(3))
print("Top 3 highest rated products:\n",df.sort_values("rating",ascending=False).head(3))
print("Sorting products by stock in descending order:\n",df.sort_values("stock",ascending=False))
df["value_score"]=df["rating"]* df["price"]
print(df)
```

```python
print("Best value product:",df.sort_values("value_score",ascending=False).head(1))
```
